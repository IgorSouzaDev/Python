```python
import pandas as pd
import matplotlib.pyplot as plt
from importlib import reload
import numpy as np
import os
```

```python
#Importar dados
data= pd.read_excel ("C:\\Users\\Igor\\Documents\\Dataframe\\Vendas\Vendas.xlsx")
display(data)
```

```python
#Faturamento total
faturamento_total = data["Valor Final"].sum()
print(faturamento_total)
```

```python
cont=data["Quantidade"].sum()
cont
```

```python
#Faturamento por Produto
Produt = data[["Produto","Quantidade"]].groupby("Produto").sum()
display(Produt)
```

```python
Count = data[["ID Loja","Produto","Quantidade","Valor Final"]].groupby(["ID Loja", "Produto"]).sum()
display(Count)
```

```python
#Fazendo um novo dataframe para o gráfico de barras
df_agrupado=data.groupby('Produto',sort=False).agg({'Quantidade':np.sum}).reset_index()
display(df_agrupado)
```

```python
# Creating fig
fig = plt.figure(figsize = (10, 4))
# creating the bar plot
plt.bar(df_agrupado["Produto"],df_agrupado['Quantidade'],width=0.3,color="blue")
plt.xlim()
plt.title("Quantidade por produto")
plt.ylabel("Quantidade")
plt.xlabel("Produto")
plt.show()
```
